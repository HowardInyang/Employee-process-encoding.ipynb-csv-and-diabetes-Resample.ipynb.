```python
import pandas as pd
import scipy
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('diabetes.csv')
```

```python
data
```

Separation data for the Features (X) and Outcomes (Y) subsets

```python
X = data.drop('Outcome', axis=1)
y = data['Outcome']
```

```python
from collections import Counter
```

```python
fig, axes = plt.subplots(figsize=(5, 5))

axes.bar(Counter(y).keys(), Counter(y).values(), color='b')
axes.set_title('Original Distribution')
plt.show()
```

Data Upsampling

```python
from imblearn.over_sampling import RandomOverSampler
```

```python
ros = RandomOverSampler(sampling_strategy='minority')
X_resampled, y_resampled = ros.fit_resample(X, y)
```

```python
fig, axes = plt.subplots(figsize=(5, 5))

axes.bar(Counter(y_resampled).keys(), Counter(y_resampled).values(), color='g')
axes.set_title('After Upsampling')
plt.show()
```

Data Downsampling

```python
from imblearn.under_sampling import RandomUnderSampler

rus = RandomUnderSampler(sampling_strategy='majority')
X_resampled_down, y_resampled_down = rus.fit_resample(X, y)
```

```python
fig, axes = plt.subplots(figsize=(5, 5))
axes.bar(Counter(y_resampled_down).keys(), Counter(y_resampled_down).values(), color='r')
axes.set_title('After Downsampling')
plt.show()
```

Outcomes in one image

```python

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

axes[0].bar(Counter(y).keys(), Counter(y).values(), color='b')
axes[0].set_title('Original Distribution')

axes[1].bar(Counter(y_resampled).keys(), Counter(y_resampled).values(), color='g')
axes[1].set_title('After Upsampling')

axes[2].bar(Counter(y_resampled_down).keys(), Counter(y_resampled_down).values(), color='r')
axes[2].set_title('After Downsampling')

plt.tight_layout()
plt.show()
```
